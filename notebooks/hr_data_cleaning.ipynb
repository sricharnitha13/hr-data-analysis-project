```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,5))
sns.countplot(y=data['Department'], order=data['Department'].value_counts().index)
plt.title('Employee Distribution by Department')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='Department', y='Salary', data=data)
plt.xticks(rotation=45)
plt.title('Salary Distribution by Department')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Corrected code
sns.countplot(y=data['TermReason'], 
              order=data['TermReason'].value_counts().index, 
              hue=data['TermReason'],  # Assign `hue`
              palette="Reds_r", 
              legend=False)  # Disable legend since `hue` is same as `y`

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Count terminations by department
termination_counts = data[data["EmploymentStatus"].isin(["Resigned", "Fired"])] \
                     .groupby("Department")["EmpID"].count() \
                     .reset_index().sort_values(by="EmpID", ascending=False)

# Plot the termination count
plt.figure(figsize=(10,5))
sns.barplot(y="Department", x="EmpID", data=termination_counts, hue="Department", palette="Reds_r", legend=False)
plt.xlabel("Number of Terminations")
plt.ylabel("Department")
plt.title("Terminations per Department")
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
sns.boxplot(x="Department", y="Salary", data=data, hue="Department", palette="Blues", legend=False)
plt.xlabel("Department")
plt.ylabel("Salary")
plt.title("Salary Distribution Across Departments")
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.boxplot(x="PerformanceScore", y="Absences", data=data, hue="PerformanceScore", palette="coolwarm", legend=False)
plt.xlabel("Performance Score")
plt.ylabel("Absences")
plt.title("Absences vs Performance Score")
plt.show()
```
